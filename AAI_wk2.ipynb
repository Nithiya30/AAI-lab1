```python
import numpy as np
```

```python
def step_function(x):
    return 1 if x >= 0 else 0
def train_perceptron(X, T, lr, gate_name):
    w = np.zeros(X.shape[1])
    b = 0
    epoch = 0

    print(f"\n===== Training {gate_name} Gate =====")
    print("Initial Weights:", w)
    print("Initial Bias:", b)

    while True:
        epoch += 1
        total_error = 0
        print(f"\nEpoch {epoch}")

        for i in range(len(X)):
            yin = np.dot(X[i], w) + b
            y = step_function(yin)
            error = T[i] - y

            total_error += abs(error)

            # Weight & bias update
            w = w + lr * error * X[i]
            b = b + lr * error

            print(f"Input: {X[i]}, Target: {T[i]}, Output: {y}")
            print(f"Weights: {w}, Bias: {b}")

        if total_error == 0:
            print(f"\n{gate_name} Gate Training Completed Successfully!")
            break

    return w, b
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

T_AND = np.array([0, 0, 0, 1])
T_OR  = np.array([0, 1, 1, 1])

lr = float(input("Enter learning rate: "))

w_and, b_and = train_perceptron(X, T_AND, lr, "AND")
print("\nFinal AND Weights:", w_and)
print("Final AND Bias:", b_and)

w_or, b_or = train_perceptron(X, T_OR, lr, "OR")
print("\nFinal OR Weights:", w_or)
print("Final OR Bias:", b_or)
```
